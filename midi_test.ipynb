```python
mid1 = mido.MidiFile()
mid1.ticks_per_beat = 70
track1 = mido.MidiTrack()
mid1.tracks.append(track1)

for i, track in enumerate(mid.tracks):
    for msg in track:
        if msg.type in ['note_on', 'note_off']:
            m = mido.Message(msg.type, note=msg.note, velocity=msg.velocity, time=msg.time)
            
            track1.append(m)

mid1.save('new_song.mid')
```
